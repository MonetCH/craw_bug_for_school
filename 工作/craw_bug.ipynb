### 抓取所有學系

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
#抓取所有學系
domain =''
url = 'https://www.com.tw/cross/university_150_107.html'
page = requests.get(url)
# print(page.text)
soup = BeautifulSoup(page.text,'lxml')
links = soup.select("tr div[id^=university_dep_row_height] a")
print(soup)
list = []
for s in links:
    list.append(s.text)
    
print(len(list))

for i in range(0,45,3):
    print(list[i])
```

```python

list_url = []
domain = 'https://www.com.tw/cross/'
for url in links:
    list_url.append(url.get('href'))
    
list_url_all = []
for i in range(0,45,3):
    print(list[i]+": "+ domain + list_url[i])
    list_url_all.append(list_url[i])
       
```

```python
total_dff = []
for i in list_url_all:
    http_url = domain + i
    print(http_url)
    df = pd.read_html(http_url)
    Test_area = df[2][2]
    name = df[2][3]
    dep = df[2][4]

    # name[3:]
    total = pd.concat([Test_area,name,dep],axis=1)
    total.columns = total.loc[1]
    total_drop = total.drop([0,1,2])
    
    total_dff.append(total_drop)
    
    print(total_drop)
# total_dff.to_csv("C:\\Users\\user\\Desktop\\text\\111.csv")
```

```python
word = []
import io
for i in list_url_all:
    http_url = domain + i
    url=http_url
    response = requests.get(url)
    f = io.StringIO(response.text)
    dfs = pd.read_html(f)
    word.append(dfs)
    print(http_url +"完成")
```
